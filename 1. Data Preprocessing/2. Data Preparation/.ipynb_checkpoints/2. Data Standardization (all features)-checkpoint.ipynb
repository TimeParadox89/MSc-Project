# Data Standardization

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
```

## 1. Load the Data

```python
patients = pd.read_csv("../../Data/rounded_patients.csv", index_col=0)
```

```python
patients.head()
```

## 2. Standardize the patients

```python
std_patients = pd.DataFrame(StandardScaler().fit_transform(patients.iloc[:,0:patients.shape[1]-1]), columns=patients.columns.values[:patients.shape[1]-1])
```

## 3. Append the classification column and all the categorical features

```python
std_patients['Class'] = patients['Class']
```

```python
std_patients
```

## 4. Save to File

```python
std_patients.to_csv("../../Data/standardized_patients.csv")
```
