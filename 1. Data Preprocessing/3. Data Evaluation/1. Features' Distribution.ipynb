# <center><b><h1>Distribution of the Features </h1></b></center>

```python
import pandas as pd
import seaborn as sns
sns.set(color_codes=True)
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
from matplotlib.figure import Figure
from pylab import savefig
```

## 1. Load the Data

```python
patients = pd.read_csv("../Data/standardized_patients.csv", index_col=0, decimal=",")
patients.shape
```

as float because it loads them as object

```python
patients = patients.astype(float)#.round(2)
```

```python
patients.head()
```

## 2. Split the patients through Classes

```python
class_0 = patients[patients['Class'] == 0].copy()
class_1 = patients[patients['Class'] == 1].copy()
```

## 3. Plot the Distribution

```python
fig, ax = plt.subplots(figsize=(11.69,8.27),ncols=2, nrows=2)
row = 0
col = 0
pag = 0
with PdfPages('features_distributions.pdf') as pdf_pages:
    for column in patients.columns.values:
        ax[row%2][col%2].set_title(column)
        
        #Class 1
        sns.distplot(class_0[column].values, ax=ax[row%2][col%2], label="No-Fracture (0)")
        
        #Class 2
        countplot = sns.distplot(class_1[column].values, ax=ax[row%2][col%2], label="Fracture (+1)")
        countplot.legend()
        
        print(column)
        col = col+1
        if col%2 == 0:
            row = row+1
        

        if col%2 == 0 and row%2==0:
            pdf_pages.savefig(fig, papertype = 'a4', orientation = 'portrait')
            fig, ax = plt.subplots(figsize=(11.69,8.27),ncols=2, nrows=2)
            
    pdf_pages.savefig(fig, papertype = 'a4', orientation = 'portrait')
```
