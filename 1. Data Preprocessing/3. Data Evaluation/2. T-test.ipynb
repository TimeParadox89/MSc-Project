```python
import numpy as np
from scipy import stats
import pandas as pd
```

# 1. Prepare the data

```python
patients = pd.read_csv("../Data/standardized_patients.csv", index_col=0, decimal=",")
patients = patients.astype(float)
patients.head()
```

```python
class_0 = patients[patients['Class'] == 1].copy()
class_1 = patients[patients['Class'] == 0].copy()
```

```python
patients_stat = pd.DataFrame(columns=patients.columns.values[:-1])

patients_stat.at['t(ùëãùëñ)', :] = 0
patients_stat.at['p-value', :] = 0
```

```python
for column in patients.columns.values[:-1]:
    patients_stat.at['t(ùëãùëñ)', column], patients_stat.at['p-value', column] = stats.ks_2samp(class_0[column], class_1[column])
```

```python
patients_stat = patients_stat.T.sort_values(by=['p-value'], ascending=True)
```
